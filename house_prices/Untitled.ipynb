```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os 
```

```python
data = pd.read_csv("./kc_house_data.csv", parse_dates=['date'])
```

```python
data['sale_yr'] = data.date.dt.year
```

```python
data['sale_month'] = data.date.dt.month
```

```python
data['sale_day'] = data.date.dt.day
```

```python
data.drop(columns='date', inplace=True)
```

```python
data.drop(columns='id', inplace=True)
```

```python
data.describe()
```

```python
y_data = data['price']
y_data.head()
```

```python
x_data = data.drop(columns='price')
x_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
mean = x_data.mean()
std = x_data.std(ddof=0)
x_normed = (x_data - mean) / std
x_normed.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_normed, y_data, test_size=0.3)

print("Size of training set: ", len(x_train))
print("Size of testing set: ", len(x_test))
```

```python
arr_x_train = np.array(x_train)
arr_x_test = np.array(x_test)
arr_y_train = np.array(y_train)
arr_y_test = np.array(y_test)
```

```python
import keras as k
```

```python
def basic_model_1(x_size, y_size):
    model = k.models.Sequential()
    model.add(k.layers.Dense(128, activation='tanh', input_shape=(x_size,)))
    model.add(k.layers.Dense(64, activation='relu'))
    model.add(k.layers.Dense(y_size))
    model.compile(loss='mean_squared_error', optimizer=k.optimizers.Adam(), metrics=[k.metrics.mae])
    return model
```

```python
model = basic_model_1(arr_x_train.shape[1], 1)
model.summary()
```

```python
keras_callbacks = [
    k.callbacks.TensorBoard(log_dir='/tmp/keras_logs/model_1', histogram_freq=0,
                            write_graph=True, write_images=True, embeddings_freq=0,
                           embeddings_layer_names=None, embeddings_metadata=None),
    k.callbacks.EarlyStopping(monitor='val_mean_absolute_error', patience=20, verbose=1)
]

epochs = 500
batch_size = 128

history = model.fit(arr_x_train, arr_y_train,
                   batch_size=batch_size,
                   epochs=epochs,
                   shuffle=True,
                   verbose=2,
                   validation_data=(arr_x_test, arr_y_test),
                   callbacks=keras_callbacks)

```
