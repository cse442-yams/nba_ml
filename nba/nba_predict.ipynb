```python
import pandas as pd
import numpy as np
import os
```

```python
def concat_gamelogs():
    files = [f for f in os.listdir("data/") if f.startswith("players_gamelog")]
    df = pd.concat([pd.read_csv("data/{}".format(f), parse_dates=["GAME_DATE"]) for f in files], ignore_index=True)
    return df
```

```python
df = concat_gamelogs()
```

```python
team_groups = df.groupby(['TEAM_ID', 'TEAM_ABBREVIATION'])
```

```python
team_dict = {code: id for id, code in list(team_groups.groups.keys())}
team_dict
```

```python
def parse_matchup(matchup_str):
    my_t, kind, other_t = matchup_str.split()
    is_home = kind == "vs."
    other_id = team_dict[other_t]
    return is_home, other_id
```

```python
df['is_home'] = df.MATCHUP.map(lambda m: parse_matchup(m)[0])
```

```python
df['opp_team_id'] = df.MATCHUP.map(lambda m: parse_matchup(m)[1])
```

```python
df = df.drop(columns=['TEAM_NAME', 'VIDEO_AVAILABLE'])
```
